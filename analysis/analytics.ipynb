```python
%sql

--Média Mensal dos taxis amarelos
SELECT 
    CONCAT('R$ ', FORMAT_NUMBER(AVG(vlr_total), 2)) as media_valor_total
FROM `refined-zone`.tb_corrida_taxi_unificado
WHERE origem_taxi = 'taxi_amarelo'
```

```python
%sql
-- Media de passageiros por hora arredondando.
SELECT 
    HOUR(dt_hr_inicio) as hora_do_dia,
    CAST(AVG(qtd_pessoas) AS INTEGER) as media_passageiros,
    COUNT(*) as total_corridas
FROM `refined-zone`.tb_corrida_taxi_unificado 
WHERE MONTH(dt_hr_inicio) = 5 
    AND YEAR(dt_hr_inicio) = 2023 
GROUP BY HOUR(dt_hr_inicio) 
ORDER BY hora_do_dia;
```
